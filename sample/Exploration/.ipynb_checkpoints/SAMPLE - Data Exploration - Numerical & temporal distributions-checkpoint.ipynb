# Utils

```python
# TASK: Create subfolders for saving graphs
saving_folder_name = "Distribution analyses"
num_saving_folder = "Numerical distributions"
temp_saving_folder = "Temporal distributions"

saving_folder_path = path.join(folder_path, "..", "..", saving_folder_name) #path.join(folder_path, "..", saving_folder_name)

# 0: Create the folders   
if not path.exists(path.join(saving_folder_path, num_saving_folder, subfolder)):
    makedirs(path.join(saving_folder_path, num_saving_folder, subfolder))
    print(f"A new saving folder has been created: {saving_folder_path}/{num_saving_folder}")
                   
if not path.exists(path.join(saving_folder_path, temp_saving_folder, subfolder)):
    makedirs(path.join(saving_folder_path, temp_saving_folder, subfolder))
    print(f"\nA new saving folder has been created: {saving_folder_path}/{temp_saving_folder}")
```

```python
# ---------------------------------------------------------------
# ------------------------ FUNCTIONS ----------------------------
# ---------------------------------------------------------------

# ---------------------------------------------------------------
# TASK: Improve readiability of column names
# ---------------------------------------------------------------
def improve_col_readability(column_name):
    # Retrieve the two parts
    col_name = column_name.split("(")[0].split("[")[0].rstrip()
    unit_measu = column_name.split("(")[-1].strip(")]")
    
    # Attach suffix of duplicate columns to the coloum name
    suff_secondary_col = column_name.split("_")
    if len(suff_secondary_col) > 1:
        unit_measu = unit_measu[:-3].rstrip(")]_")
        col_name += suff_secondary_col[1]
        
    # Return an improved combination    
    simplified_col_name = r"$\bf{" + col_name.replace(' ','\\ ') + "}$ [" + unit_measu.upper() + "]"
    return simplified_col_name

# ---------------------------------------------------------------
# TASK: Prepare numerical columns and generate the graphical space
# ---------------------------------------------------------------
def filter_columns(dataset, inv_name):
    # FILTER DATA: Select only the numeric columns 
    numeric_df = dataset.select_dtypes(include = np.number)
    
    # FILTER DATA: Discard the empty columns
    column_names = [column for column in numeric_df.columns if not numeric_df[column].isnull().values.all()]
    
    # Sort columns according to the unit measure
    column_names.sort(key = lambda column_name: (
            column_name.split("(")[1].split(")")[0], # unit measure
            column_name.split("(")[0].split("[")[0].rstrip() # column name
        )) 

    # Highlight discarded columns
    discarded_columns = set(dataset.columns) - set(column_names)
    
    if len(discarded_columns) > 1: # Other columns apart from the datetime
        print(f"WARNING: [{system_name.upper()}]:{inv_name}")
        print(f"Some columns ({len(discarded_columns)}) have been discarded since they "
              f"were non-numerical values or empty columns. \n({', '.join(sorted(discarded_columns))})\n")
   
    return numeric_df, column_names

def prep_visual_env(dataset, inv_name):
    #global system_name
    sns.set_theme(style="whitegrid")
    
    # Select a meaningful subset of columns
    numeric_df, column_names = filter_columns(dataset, inv_name)
    
    # Compute the number of columns and rows of the figure 
    n_columns = 2 if len(column_names) <= 12 else 3
    raw_nRows = len(column_names)/n_columns
    n_rows = ceil(raw_nRows)
    
    # Create the figure
    fig, axes = plt.subplots(n_rows, n_columns, figsize=(10 * n_columns, 3 * n_rows), tight_layout= {"pad": 3})
    fig.suptitle(r"$\bf{" + system_name.upper().replace(' ','\\ ') + "}$: " + inv_name.capitalize(), 
                 fontsize=40 * n_columns, color='dimgray', y = 0.96) #, y=1
    
    # Make invisible the last graph in extra subplots
    extra_subplots = (axes.shape[0] * axes.shape[1]) - len(column_names)
    if extra_subplots > 0:
        for row in range(1, extra_subplots + 1):
            axes[-row, -1].set_visible(False)
    
    return fig, axes, numeric_df, column_names

# ---------------------------------------------------------------
# TASK: Carry out an analysis on the numerical distributions
# FUNCTION: Generate boxplots for the inverter data
# ---------------------------------------------------------------
def generate_boxplot(dataset, inv_name, visual_output=True):
    # Load the preparation steps (e.g., filter columns, generate the visual structure)
    fig, axes, numeric_df, column_names = prep_visual_env(dataset, inv_name)
    
    # Plot the columns
    idk_row = 0
    idk_column = 0
    for column in column_names:  
        sns.boxplot(x = column, data = numeric_df, ax=axes[idk_row, idk_column], 
                    palette="hls",  linewidth=2, 
                    flierprops = dict(marker='o', markersize=6, 
                                      markerfacecolor='tab:red', 
                                      markeredgecolor='dimgrey',markeredgewidth=1.5))
        # Set graphical parameters
        axes[idk_row, idk_column].tick_params(axis='x', which='major', labelsize=14)
        axes[idk_row, idk_column].set_xlabel(improve_col_readability(column), fontsize=20)
 
        axes[idk_row, idk_column].ticklabel_format(axis="x", useMathText=True) # style="plain"
        axes[idk_row, idk_column].xaxis.offsetText.set_fontsize(14)
        
        axes[idk_row, idk_column].xaxis.set_major_locator(MaxNLocator(integer=True))
                             
        # Change row/column counters
        if idk_row + 1 < axes.shape[0]:
            idk_row += 1
        else:
            idk_row = 0
            idk_column += 1
        
    # Saving the graphical panels as a picture (png)
    name_to_save = "numDist_" + system_name.replace(" ", "") + "_" + inv_name + + "_" + subfolder + ".png"
    path_to_save = path.join(saving_folder_path, num_saving_folder, subfolder, name_to_save)
    fig.savefig(path_to_save)
    
    # Visual verbosity 
    if visual_output:
        plt.show()
    else:
        print(f"A boxplot for '{inv_name}' has been saved (visual output = False)")
    plt.close()
    
# ---------------------------------------------------------------
# TASK: Carry out an analysis on the temporal distributions
# FUNCTION: Generate histograms for the inverter data
# --------------------------------------------------------------
def generate_histograms(dataset, inv_name, visual_output=True):
     # Load the preparation steps (e.g., filter columns, generate the visual structure)
    fig, axes, numeric_df, column_names = prep_visual_env(dataset, inv_name)
    
    # Save dates (used as data for the x axis)
    dates = dataset.iloc[:, 0]
    num_days = (dates.iloc[-1] - dates.iloc[0]).days
    
    # Increase fig height to improve readability of histograms
    fig.set_figheight(4 * axes.shape[0])
    
    # Plot the columns
    idk_row = 0
    idk_column = 0
    for column in column_names:  
        # Generate histogram
        sns.histplot(x = dates, y = numeric_df[column], ax=axes[idk_row, idk_column], 
                     bins = num_days // 7, cbar=True, color="orange", cbar_kws={"shrink": 0.7})
        
        # Set subplot title
        full_col_name = improve_col_readability(column).split("[")
        axes[idk_row, idk_column].set_title(full_col_name[0], fontsize=35, color= "dimgrey", pad=10)
        
        # Set graphical parameters - axis labels
        axes[idk_row, idk_column].set_ylabel("["+full_col_name[1], fontsize="large", fontweight='bold',) 
        axes[idk_row, idk_column].set_xlabel("DAYS", fontsize="x-large", fontweight='bold', labelpad = 15)
        
        # Set graphical parameters - axis ticks
        axes[idk_row, idk_column].tick_params(axis='x', which='major', labelsize=12, grid_linestyle = "-.")
        axes[idk_row, idk_column].xaxis.set_major_formatter(DateFormatter("%Y-%m"))
        axes[idk_row, idk_column].xaxis.set_major_locator(MaxNLocator(8)) #MonthLocator(interval=4)
        
        # Scientific notation label
        axes[idk_row, idk_column].ticklabel_format(axis="y", useMathText=True) # style="plain"
        axes[idk_row, idk_column].yaxis.offsetText.set_position((-0.10, 0))
        axes[idk_row, idk_column].yaxis.offsetText.set_fontsize(15)
        axes[idk_row, idk_column].yaxis.offsetText.set_fontweight("bold")
              
        # Change row/column counters for the next subplot
        if idk_row + 1 < axes.shape[0]:
            idk_row += 1
        else:
            idk_row = 0
            idk_column += 1
        
    # Saving the graphical panels as a picture (png)
    name_to_save = "tempDist_" + system_name.replace(" ", "") + "_" + inv_name + "_" + subfolder + ".png"
    path_to_save = path.join(saving_folder_path, temp_saving_folder, subfolder, name_to_save)
    fig.savefig(path_to_save)
    
    # Visual verbosity 
    if visual_output:
        plt.show()
    else:
        print(f"A histogram for '{inv_name}' has been saved (visual output = False)")
    plt.close()
    
# ---------------------------------------------------------------
# TASK: Carry out an analysis of the correlation coefficients 
# FUNCTION: Generate a visualization (i.e., heatmap) for the correlation coef.
# --------------------------------------------------------------
def compute_correlationMatrices(dataset, inv_names, batch_id=None):
    n_columns = 2 if len(inv_names) != 1 else 1
    n_rows = ceil(len(inv_names)/n_columns)

    # Create the figure
    fig, axes = plt.subplots(n_rows, n_columns, squeeze=False,
                             figsize=(15 * n_columns, 10 *n_rows), # 15 o 30 
                             tight_layout= {"w_pad": 4, "h_pad":3})
    dynamic_title = "Correlation Matrices" if len(inv_names) > 1 else "Correlation Matrix"
    fig.suptitle(r"$\bf{" + system_name.upper().replace(' ','\\ ') + "}$: " + dynamic_title, 
                 fontsize=70, color='dimgray', y=1) 

    # Make invisible the last graph in extra subplots
    if n_rows > 1:
        extra_subplots = (axes.shape[0] * axes.shape[1]) - len(inv_names)
        if extra_subplots > 0:
            for row in range(1, extra_subplots + 1):
                axes[-row, -1].set_visible(False)

    idk_row = 0
    idk_column = 0
    for inv_name in inv_names:
        
        # Get the columns filtered 
        numeric_df, column_names = filter_columns(dataset[inv_name], inv_name)

        # Compute correlation matrix (Pearson correlation coefficient)
        corrMatrix = numeric_df[column_names].corr()
        
        # Remove nan columns/rows after performing the correlation coefficient 
        corrMatrix.dropna(axis=0, how="all", inplace=True)
        corrMatrix.dropna(axis=1, how="all", inplace=True)
        
        sns.heatmap(corrMatrix.round(decimals=2), ax = axes[idk_row, idk_column],
                    annot=True, fmt =".1g", cmap = "coolwarm",
                    center = 0, vmin = -1, vmax=1,  cbar_kws={"shrink": .4, "pad":-0.05, "anchor": (0.0, 0.7)},
                    linewidths = 1,
                    mask = np.triu(corrMatrix, k = 1))
        
        cbar = axes[idk_row, idk_column].collections[0].colorbar
        cbar.ax.set_ylabel("Correlation coefficient", fontsize=20, labelpad = -80, color='dimgray')
        cbar.ax.tick_params(axis='y', width=2, length=5, direction = "inout", color="white")
        
        # Set title
        axes[idk_row, idk_column].set_title(inv_name.capitalize(), fontsize=55, color= "dimgrey",
                                            pad=10, y = 0.6, x = 0.70)
        
        # Graphical settings: Y axis
        simplified_y_ticks = [improve_col_readability(label.get_text()) for label in axes[idk_row, idk_column].get_yticklabels()]
        axes[idk_row, idk_column].set_yticklabels(simplified_y_ticks, size=14)
        
        # Graphical settings: X axis
        simplified_x_ticks = [improve_col_readability(label.get_text()).split("[")[0] for label in axes[idk_row, idk_column].get_xticklabels()]
        axes[idk_row, idk_column].set_xticklabels(simplified_x_ticks, size=12, rotation = 30,
                                                  rotation_mode="anchor",  ha="right", va="top")
        
        # Change row/column counters
        if idk_column + 1 < axes.shape[1]:
            idk_column += 1
        else:
            idk_column = 0
            idk_row += 1

    # Save the correlation matrices/matrix
    if batch_id != None:
        labels = list(ascii_uppercase) # ["A", "B", "C", "D", "E", "F", "G", "H"]
        name_to_save = "corrMatrix_StringInverter_" + labels[batch_id] + "_" + subfolder +".png"
    else:
        name_to_save = f"corrMatrix_Inverter_{subfolder}.png"
        
    path_to_save = path.join(saving_folder_path, name_to_save)
    fig.savefig(path_to_save, bbox_inches='tight', pad_inches=1)
    print(f"The correlation matrices for {system_name.upper()} have been saved to: {saving_folder_path}")

    # Visualize the correlations
    plt.gcf().text(0.820, 0.106, "Mars", fontsize=10, color="white")
    plt.show()
```

# Numerical distributions 

```python
# TASK 1A: GENERATE NUMERICAL DISTRIBUTIONS (i.e., BOX-PLOTS)
for inv_name in inv_names:
    generate_boxplot(inv_data[inv_name], inv_name)
```

# Temporal distributions 

```python
# TASK 1B: GENERATE TEMPORAL DISTRIBUTIONS (i.e., HISTOGRAMS)
for inv_name in inv_names:
    generate_histograms(inv_data[inv_name], inv_name)
```

# Correlation coefficients

```python
# TASK 1C: COMPUTE AND VISUALIZE (i.e., heatmaps) CORRELATION COEFFICIENTS of the STRING INVERTER DATA
compute_correlationMatrices(inv_data, inv_names)
```

# [ONLY FOR BINETTO 1&2] STRING INVERTERS 

```python
# [ONLY FOR BINETTO 1&2]: A) GENERATE NUMERICAL DISTRIBUTIONS (i.e., BOX-PLOTS) OF THE STRING INVERTERS
if string_inv_data:
    n_visual_output = 1
    for idk, name in enumerate(string_inv_names):
        if idk + 1 <= n_visual_output:
            generate_boxplot(string_inv_data[name], name)
        else:
            generate_boxplot(string_inv_data[name], name, visual_output = True)
else:
    print("Sorry, there is not such string inverter data, as the system is", system_name.upper())
```

```python
# [ONLY FOR BINETTO 1&2]: B) GENERATE TEMPORAL DISTRIBUTIONS (i.e., HISTOGRAMS) OF THE STRING INVERTERS
if string_inv_data:
    n_visual_output = 1
    for idk, name in enumerate(string_inv_names):
        if idk + 1 <= n_visual_output:
            generate_histograms(string_inv_data[name], name)
        else:
            generate_histograms(string_inv_data[name], name, visual_output = True)
else:
    print("Sorry, there is not such string inverter data, as the system is", system_name.upper())
```

```python
# -----------------------------------------------------------------------
# ----- [ONLY FOR BINETTO 1&2]: C) COMPUTE CORRELATION COEFFICIENTS -------- 
# -----------------------------------------------------------------------
if string_inv_data:
    # Decide of the organize the graphical panel
    max_num_subplots = 4
    num_ranges = len(string_inv_names)/max_num_subplots
    
    # Create indices accordinglly 
    idks= np.arange(len(string_inv_names))
    ranges = np.array_split(idks,num_ranges)
    
    for k_batch, idk_range in enumerate(ranges):
        stringInv_to_analyse = [string_inv_names[idk] for idk in idk_range]
        compute_correlationMatrices(string_inv_data, stringInv_to_analyse, batch_id = k_batch)
else:
    print("Sorry, there is not such string inverter data, as the system is", system_name.upper())
```
