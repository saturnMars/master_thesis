# Read the files

```python
folder_name = 'Grid search - Generalized Version'
reading_path = path.join('data', system_name.upper(), system_name.upper(), 'UC2 - LSTM', folder_name) 
```

```python
load_merged_inv_data = True
if load_merged_inv_data:
    file_names = ['All inverters']
else:
    file_names = ['INV1', 'INV2', 'INV3', 'INV4']
```

```python
print("FOLDER:", folder_name, "\n")
inv_perfomance = dict()
for inv_name in file_names:
    
    file_name = f'lstm_gridSearch_perfomance_{inv_name}.txt'
    file_path = path.join(reading_path, file_name)
    
    if not path.exists(file_path):
        print(f"{inv_name}: File not found!")
        continue
    
    config_perfomance = []
    with open(file_path, 'r') as file:
        raw_text = file.read().split('\n--------------------------------------------------\n')
        raw_text = [config.split('\n') for config in raw_text]
        
        for config in raw_text:
            perfomance = dict()
            for metrics in config:
                metrics_name = metrics.split(':')[0].strip("' ")
                metrics_value = metrics.split(':')[-1].strip("' ")

                try: 
                    if metrics_value.isdigit(): 
                        metrics_value = int(metrics_value) 
                    else: 
                        metrics_value = float(metrics_value) 
                    perfomance[metrics_name] = metrics_value
                    
                except ValueError:
                    continue
            if perfomance:
                config_perfomance.append(perfomance)

    inv_perfomance[inv_name] = config_perfomance
    print(f"{inv_name}: File loaded with ({len(config_perfomance)} configs)")
```

## Compute the f1 scores for each clas

```python
compute_f1_score = lambda precision, recall: 2 * ((precision * recall) / (precision + recall))
for inv_name in inv_perfomance.keys(): 
    all_perfomance = inv_perfomance[inv_name]
    
    enhanced_all_perfomance = []
    for config_performance in all_perfomance:
        
        class_labels = defaultdict(list)
        for metrics_name in config_performance.keys():
            if 'VAL' in metrics_name:
                if metrics_name.split(' ')[-1].strip() in ascii_uppercase:
                    class_labels[metrics_name.split(' ')[-1].strip()].append(metrics_name) 
        f1_scores = []
        for class_letter, (precision_name, recall_name) in class_labels.items():     
            try: 
                f1_score = compute_f1_score(config_performance[precision_name], config_performance[recall_name])
                f1_scores.append(f1_score)
            except ZeroDivisionError:
                f1_score = 0
            config_performance[f'VAL F1 SCORE {class_letter}'] = np.round(f1_score, 4)
            
        # Compute the average value
        if len(f1_scores) > 0:
            avg_f1_score = np.round(np.mean(f1_scores), 4)
        else:
            avg_f1_score = 0
        config_performance[f'VAL F1 SCORE [available classes]'] = avg_f1_score
        
        # Sort the metrics
        enhanced_all_perfomance.append(dict(sorted(config_performance.items(), 
                                                   key = lambda item: (item[0].split(' ')[-1], item[0]), reverse = True)))
    inv_perfomance[inv_name] = enhanced_all_perfomance
    
    print(f"{inv_name}: the F1 scores for each class have been computed")
```

# Visualize

```python
# Create the main figure
pov_elev = 20
pov_angle = -120

plt.style.use('seaborn-bright')

# Create the subplots
for idk, inv_name in enumerate(inv_perfomance.keys()):
    
    # Get the perfomance
    configs_perfomance = inv_perfomance[inv_name]
    
    # Create the data for the graph
    x_data = [config['WINDOW DIMENSIONS'] for config in configs_perfomance] 
    y_data = [config['STARTING NEURONS'] for config in configs_perfomance]

    # A column for each metrics
    metrics_names = [metrics_name for metrics_name in configs_perfomance[0].keys() 
                     if 'VAL' in metrics_name and 'CATEGORICAL' not in metrics_name]
    num_rows = 3 if len(metrics_names) <= 6 else 4
    num_columns = ceil(len(metrics_names) / num_rows)
        
    fig = plt.figure(figsize=(13 * num_columns, 13 * num_rows), facecolor = "white")
    title = r"$\bf{" + "LSTM - Hyperparameter optimization".replace(' ','\\ ') + "}$"
    title += " [Validation set]"
    title += f"\n{system_name}: "
    
    if load_merged_inv_data: 
        title += inv_name
    else:
        title += f"Inverter N°{inv_name[-1]}"
    fig.suptitle(title, fontsize = 100, color = 'firebrick', y = 1.01)
    
    for idk_plot, metric_name in enumerate(metrics_names):
        
        simplified_name =  r"$\bf{" + metric_name.replace('VAL', "").lstrip().replace(' ', '\\ ') + "}$" # "[Validation]"
        
        check_class_names = [class_metrics_name for class_metrics_name in ['PRECISION', 'RECALL', 'F1 SCORE'] 
                             if class_metrics_name in metric_name.upper()]
        
        if len(check_class_names) > 0:
            splitted_name = simplified_name.split(' ')
            letter = splitted_name[-1].strip('}$')
            
            if letter in ascii_uppercase:
                simplified_name = simplified_name.replace(f' {letter}', f' [Alarm\\ {letter}]') 
                simplified_name += f"\n'{class_names[letter]}'"
            
        # Get the metrics values
        z_data = np.array([config[metric_name] for config in configs_perfomance])
        
        if 'LOSS' not in metric_name.upper():
            z_data = z_data * 100
        
        # Ad the subplots
        main_ax = fig.add_subplot(num_rows, num_columns, idk_plot + 1, projection='3d', computed_zorder=False)

        # GRAPH: Line 1
        surf = main_ax.plot_trisurf(x_data, y_data, z_data, cmap = 'Reds', shade = True, linewidth = 1, edgecolor = 'white', 
                                    antialiased = True,  lightsource = LightSource(azdeg = 150, altdeg = 45), zorder = 0) 
        # Fix a library bug
        surf._facecolors2d = surf._facecolor3d
        surf._edgecolors2d = surf._edgecolor3d

        # 3D: Point of view
        main_ax.view_init(elev = pov_elev, azim = pov_angle)
        
        # GRAPH: Maxiumum peak
        if 'LOSS' in metric_name.upper():
            max_peak = np.argmin(z_data)
            peak = np.min(z_data)
        else:
            max_peak = np.argmax(z_data)
            peak = np.max(z_data)
            
        if np.std(z_data) != 0:
            main_ax.scatter(x_data[max_peak], y_data[max_peak], z_data[max_peak], 
                            label = f"{y_data[max_peak]} neurons with {x_data[max_peak]} hours "\
                                    f"({np.round(peak, 1)} {' %' if 'LOSS' not in metric_name.upper() else ''})",
                            linewidths = 5, facecolors = 'white', edgecolors='black', s = 2000, alpha = 0.7, zorder = 1)
        
        main_ax.set_title(simplified_name, fontsize = 45, y = 0.97)
        main_ax.set_xlabel('Temporal window (hours)', fontsize = 30, labelpad = 20)
        main_ax.set_ylabel('Starting neurons', fontsize = 30, labelpad = 20)
        
        if np.std(z_data) != 0:
            title_name = r"$\bf{" + ('Minimum'  if 'LOSS' in metric_name.upper() else 'Maximum') + '\\ peak' + "}$"
            main_ax.legend(loc = 'center right', 
                           bbox_to_anchor = (0.81, (0.70 if 'LOSS' in metric_name.upper() else 0.29)),
                           shadow = True,
                           fontsize = 21,  markerscale = 0.4, title = title_name ,title_fontsize = 22, 
                           framealpha = 0.9, facecolor = 'white', edgecolor = 'firebrick')\
            .get_frame().set_boxstyle('Round', pad = 0.3, rounding_size = 1)
        else:
            if np.mean(z_data) == 0:
                handle = Patch(facecolor = 'seashell', edgecolor = 'Black', label = 'All zero values')
                main_ax.legend(handles = [handle], loc='best', bbox_to_anchor = (0.58, 0.76), fontsize = 21, shadow = True, 
                               framealpha = 1, facecolor = 'white', edgecolor = 'black')\
                .get_frame().set_boxstyle('Round', pad = 0.3, rounding_size = 1)
                
        x_labels = np.unique(x_data)
        main_ax.set_xticks(x_labels, minor = True)
        
        x_labels = np.delete(x_labels, [1, 3]) 
        main_ax.set_xticks(x_labels, minor = False)
        main_ax.set_yticks(np.unique(y_data))
        
        if 'LOSS' not in metric_name.upper():
            main_ax.set_zlabel("Percentage (%)", fontsize = 30, labelpad = 30 if np.std(z_data) >= 1 else 35, rotation = -60)
            main_ax.zaxis.set_major_locator(MaxNLocator(integer=True))
        
        if (0 < np.max(z_data) <= 1) or (0 < np.std(z_data) <= 0.1):
            order = int(np.abs(np.round(np.log10(np.std(z_data)))))
            main_ax.zaxis.set_major_locator(MaxNLocator(((order + 1) * 2)))
            main_ax.zaxis.set_major_formatter(FormatStrFormatter(f'%.{order}f'))   
            
        if np.std(z_data) == 0:
            main_ax.zaxis.set_major_locator(MaxNLocator(1))
        
        main_ax.tick_params(axis='x', labelsize = 20, grid_linewidth = 1)
        main_ax.tick_params(axis='y', labelsize = 20, grid_linewidth = 1)
        main_ax.tick_params(axis='z', labelsize= 20, pad = 15)
        
    fig.tight_layout(h_pad = 2)
    
    # Save the figure 
    fig.savefig(path.join(reading_path, f'{inv_name}_behaviour.png'), bbox_inches='tight', pad_inches = 1)
    plt.show()
```

# Find the best config among the inverters

## Merge the perfomance of each inverter

```python
config_perfomance = defaultdict(list)
for idk, inv_name in enumerate(inv_perfomance.keys()):
    
    # Get the perfomance
    configs_perfomance = inv_perfomance[inv_name]
    
    for config in configs_perfomance:
        config_name = str(config['STARTING NEURONS']) + 'N_' + str(config['WINDOW DIMENSIONS']) + "H"
        
        metrics = {metrics_name: value for metrics_name, value in config.items() 
                   if ('VAL' in metrics_name) and ('LOSS' not in metrics_name) and ('WINDOW DIMENSIONS' not in metrics_name) 
                   and ('STARTING NEURONS' not in metrics_name) and ('BEST_EPOCH' not in metrics_name)}
        config_perfomance[config_name].append(metrics)
print(f"OK: Merged data of {len(inv_perfomance.keys())} inverters (i.e., {', '.join(inv_perfomance.keys())})")
```

## Compute the average value

```python
merged_config_perfomance = dict()
for config, inv_metrics in config_perfomance.items():

    # Merge the score
    merged_scores = defaultdict(list)
    for perfomance in inv_metrics:
        for metrics_name, score in perfomance.items():
            merged_scores[metrics_name].append(score)
    merged_config_perfomance[config] = merged_scores  
            
# Compute the average values
for config_name, metrics_list in merged_config_perfomance.items():

    for metrics in metrics_list:
        merged_config_perfomance[config_name][metrics] = np.round(np.mean(merged_config_perfomance[config_name][metrics]), 4)
print(f"Averaged perfomance of the {len(merged_config_perfomance.keys())} configurations --> OK")
```
