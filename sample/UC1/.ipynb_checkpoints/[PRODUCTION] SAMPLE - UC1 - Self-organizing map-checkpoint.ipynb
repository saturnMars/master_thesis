```python
def detrending(df, feature, daily_obs = 16, num_weeks = 2, verbose = True):

    # Visualize the possible trend
    #if verbose:
        #days = 7 * (num_weeks//2)
        #analysis = seasonal_decompose(df[feature],  model='additive', extrapolate_trend='freq', period = days * daily_obs, 
                                       #two_sided = False)
        #plt.rcParams.update({'figure.figsize': (10,10)})
        #analysis.plot(observed = False, seasonal=True, resid=False).suptitle(f'Additive Decomposition [{feature}]', fontsize=18)
        #plt.show()
    
    # 1) Compute the moving average 
    moving_average_col = feature + f"_{num_weeks}w"
    df[moving_average_col] = df[feature].rolling(window = daily_obs * 7 * num_weeks, min_periods = None).mean()
        
    # 2) Compute the detrended values
    detrend_col = feature + f"detrend_{num_weeks}w"
    detrended_values = df[feature] - df[moving_average_col]
    df[detrend_col] = detrended_values
    
    # 3) Compute the moving average over the detranded feature
    detrend_mean_col = feature + f"detrend_{num_weeks}w_mean"
    df[detrend_mean_col] = df[detrend_col].rolling(window = daily_obs * 7 * num_weeks, min_periods = 1).mean()
    
    # Visualize the moving average
    cols_to_visualize = [feature, moving_average_col, detrend_mean_col]
    if verbose: 
        colors = ["grey", "blue", "red"]
        df[cols_to_visualize].plot(color = colors, linewidth=1, figsize=(6,3))

        # Some graphical parameters
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)
        plt.title(feature, fontsize=20)
        plt.xlabel('Time', fontsize=16)
        plt.ylabel(feature, fontsize=16)
        plt.legend(labels = [feature, f'{feature}: {num_weeks}-week mean',f'{feature}: {num_weeks}-week detrend mean'], fontsize=11)
        plt.show()
    
    # Drop artefacts
    df.drop(columns = cols_to_visualize + [detrend_col], inplace=True)
    
    return detrended_values
```

## Utils

```python
def pre_processing_data(df, regression_model, standard_scaler, percent_error_threshold, daily_obs_detrending, num_weeks_detrending,
                        verbose = False):
    
    # 0) Compute the DC power
    print(40 * "-", "(0) Compute the DC Power", 40 * "-", "\n")
    # 0.1) FORMULA: P = V * A --> Power = Voltage * Ampere
    dc_power = df["Vcc 1 (V)"] * df["Cc 1 (A)"]

    # Transform the WATT (W) into KILOWATT (kW)
    watt_to_kw = lambda watt: watt/1000
    df.loc[:, 'DC Power (kW)'] = watt_to_kw(dc_power)
    print("\t\t\t\t\t[DC POWER] Finished")
    
    # A) Regression
    print(40 * "-", "(A) APPLY THE REGRESSION", 40 * "-", "\n")

    # A.2) Prepare the input data
    x_features_names = list(dict(regression_model.params).keys())
    y_feature_name = "Pac R (kW)"
    
    if any(x_features_names) not in df.columns:
        df["Irradiance (W/mq)^2"] = df['Irradiance (W/mq)'] ** 2
        df["Irradiance (W/mq)^3"] = df['Irradiance (W/mq)'] ** 3
    input_data = df[[y_feature_name] + x_features_names]
    
    # A.3) Apply the threshold 
    idk_potential_outliers = pre_utils.compute_outliers_above_threshold(regression_model, y_feature_name, input_data,
                                                                        x_features_names, percent_error_threshold, verbose)
    
    if idk_potential_outliers == df.index[-1]:
        print(f"The very last observation (i.e., {inv_df.index[-1].strftime('%d-%m-%Y, %H:%M')}) has been discarded! The warning is 0")
        return None
    
    # A.4) Drop the outliers
    df.drop(idk_potential_outliers, inplace = True)
    df.drop(columns = ["Irradiance (W/mq)^2", "Irradiance (W/mq)^3"], inplace = True)
    print("\t\t\t\t\t[REGRESSION] Finished")
    
    # B) Detrending
    print("\n", 40 * "-", "(B) APPLY THE DETRENDING", 40 * "-", "\n")
    
    # B.1) Selecting the columns
    columns_to_ignore = ['Vac (V)', 'Vac R (V)', 'Vac S (V)', 'Vac T (V)', 'Vcc 1 (V)', 'E. totale (kWh)']
    columns = [col for col in df.columns.tolist() if col not in columns_to_ignore]
    columns = sorted(columns, key = lambda col_name: col_name.split('(')[1])
    if verbose: 
        print(f"COLUMNS ({len(columns)}):", columns)
    
    # B.2) Compute the detrended values for each columns
    for col in columns:
        if verbose: 
            print("-" * 40 + f" {col} " + "-" * 40)
        
        # Compute the detrended values
        detrended_col = pre_utils.detrending(df, col, daily_obs_detrending, num_weeks_detrending, verbose)
        
        df.loc[:, col] = detrended_col
    print("\t\t\t\t\t[DETRENDING] Finished")
    
    # C) Standard scaler 
    print("\n", 40 * "-", "(C) APPLY THE STANDARD SCALER", 40 * "-", "\n")
    
    # C.2) Aply the standard scaler to each feature (i.e., column)
    for idk_feature, feature in enumerate(standard_scaler['features']):
        if feature in df.columns:
            
            # Retrieve the mean & variance of the feature
            mean = standard_scaler['mean_values'][idk_feature]
            variance = standard_scaler['variance_values'][idk_feature]
            
            # Apply the standard scaler [z = (x - mean) / std]
            stdScaler = lambda value: (value - mean)/np.sqrt(variance)
            df.loc[:, feature] = df[feature].apply(stdScaler)
            
            if verbose:
                print(f"\n---- {feature} ----")
                print("MEAN:", round(mean, 2))
                print("VARIANCE:", round(variance, 2))
                print( "-" * 20)
    print("\t\t\t\t\t[STANDARD SCALER] Finished")
    
    # D) Reorder the columns
    new_cols_order =['Iac R (A)', 'Iac S (A)', 'Iac T (A)', 'Vac R (V)', 'Vac S (V)', 'Vac T (V)', 'Pac R (kW)', 
                     'E. totale (kWh)', 'Cc 1 (A)', 'Vcc 1 (V)', 'DC Power (kW)', 'Inverter temp. (°C)', 'Irradiance (W/mq)',
                    'Amb. Temp (°C)']
    if all([col in new_cols_order for col in df.columns]):
        df = df.reindex(columns = new_cols_order)
        
    # E) Keep only the last 24 hours
    starting_ts = df.index[-1] - pd.Timedelta(23, unit= "hours")
    last_24hours = df[starting_ts:]
    print("-" * 60 + f"\nPERIOD SELECTED (last 24h)\n--> FROM '{starting_ts.strftime('%d-%m-%Y (%H:%M)')}' "\
          f"TO '{df.index[-1].strftime('%d-%m-%Y (%H:%M)')}'\n--> Hourly observations: {len(last_24hours)}\n" + "-" * 60)
    
    return last_24hours
```

## 3.C) Load the pre-trained SOMs for each inverter

```python
def load_pretrained_soms(reading_folder):
    pretrained_soms = dict()
    for inv_name in inv_names:
        print("\n" + 40 * "-", inv_name, 40 * "-" + "\n")
        
        # Skip this (and the follow-up) stage, since this timestamp has been skipped
        if inv_name in inv_warnings.keys():
            print(f"[{inv_name}] The warning level for this inverter has already been computed")
            continue

        # Saving path
        file_name = f"trained_som_{inv_name}.p"
        saving_som_folder_name = "Trained SOM"
        loading_path = path.join(reading_som_folder, saving_som_folder_name, trained_version, file_name)

        # Load the trained SOM 
        if path.exists(loading_path):      
            with open(loading_path, 'rb') as infile:
                pretrained_soms[inv_name] = load_pickle_file(infile)
                print(27 * "-", f"The trained SOM has been LOADED!" + 40 * "-")
                print(f"Version [{inv_name}]: {trained_version}\n" + 100 * "-")
        else:
            print(f"ISSUE: File not found!\n{loading_path})")
            
    return pretrained_soms
```

```python
# Set the folder for loading the pretrained SOMs
reading_som_folder = path.join(system_path, "..", "..", "SOMs")

# Load the pretrained SOMs
pretrained_soms = load_pretrained_soms(reading_som_folder)
```

# 3.D) Compute the KPI scores

```python
def compute_kpi_scores(trained_som, inv_data, sliding_window = 24):
    inv_kpi_scores = dict()
    for inv_name in trained_som.keys():
        
        # Skip this (and the follow-up) stage, since this timestamp has been skipped
        if inv_name in inv_warnings.keys():
            print(f"[{inv_name}] The warning level for this inverter has already been computed")
            continue
            
        print(40 * "-", inv_name, 40 * "-")

        # Load the trained SOM for this inverter
        som = trained_som[inv_name]

        # Load the inverter data
        df = preprocessed_last_24h_inv_data[inv_name]
        
        # Read the train neuron probabilities 
        file_name = f"train_prob_matrix_{inv_name}.csv"
        file_path = path.join(reading_som_folder,  "Trained SOM", trained_version, file_name)
        train_neuron_prob = pd.read_csv(file_path).values
        
        # Compute the KPI scores
        kpi_scores = utils.compute_kpi_scores_minimal(som, df, train_neuron_prob, sliding_window, verbose = True)
        
        # Consider only the last KPI
        num_selected_kpi = 2
        last_kpi = kpi_scores.iloc[-num_selected_kpi:]  
        inv_kpi_scores[inv_name] = last_kpi
    return inv_kpi_scores
```

```python
inv_kpi_scores = compute_kpi_scores(pretrained_soms, preprocessed_last_24h_inv_data)
```

## 3.E) Load the numerical thresholds

```python
thresholds_inv = dict()
for inv_name in inv_names:
    print("\n" + 40 * "-", inv_name, 40 * "-" + "\n")
    
    # Skip this (and the follow-up) stage, since this timestamp has been skipped
    if inv_name in inv_warnings.keys():
        print(f"[{inv_name}] The warning level for this inverter has already been computed")
        continue

    # Saving paths
    threshold_file_name = f"thresholds_{inv_name}.csv"
    loading_threshold_path = path.join(reading_som_folder,  "KPI scores", trained_version, threshold_file_name)
    
    # Load the KPI scores in the case they have already been computed previously
    if path.exists(loading_threshold_path):
        print("\t\t\t" + 45 * "-" + f"\n\t\t\tLoading the pre-computed thresholds...\n\t\t\t" + 45 * "-")
        
        # Load thresholds
        thresholds_df = pd.read_csv(loading_threshold_path)
        threshold1 = thresholds_df.iloc[0, 1]
        threshold2 = thresholds_df.iloc[1, 1]
        thresholds_inv[inv_name] = (threshold1, threshold2)
        print(f"\t\t\tFORMULA: {thresholds_df.iloc[0, 0]} --> THRESHOLD 1: {round(threshold1, 4)}")
        print(f"\t\t\tFORMULA: {thresholds_df.iloc[1, 0]} --> THRESHOLD 2: {round(threshold2, 4)}")
```

## 3.F) Generate the status/warnings

```python
for inv_name in inv_names:
    print(50 * "-", inv_name, 50 * "-")
    
    # Skip this (and the follow-up) stage, since this timestamp has been skipped
    if inv_name in inv_warnings.keys():
        print(f"[{inv_name}] The warning level for this inverter has already been computed")
        continue
    
    # Retrieve inverter data
    kpi_scores = inv_kpi_scores[inv_name]
    num_thresholds= thresholds_inv[inv_name]
    
    # Retrieve the warnings according to thier KPI scores
    generated_warnings = utils.create_warning(kpi_scores, num_thresholds,  keep_zero_values = True)
    
    # Set the maximum warning level to 2
    cond_lim_warn = generated_warnings['Warning level'] > 2
    generated_warnings.loc[cond_lim_warn, 'Warning level'] = 2
    
    # Keep only the warning of the last timestamp computed usig the 24-hour sliding window
    generated_warnings = generated_warnings.iloc[-1:]
    inv_warnings[inv_name] = generated_warnings
    
    print("-" * 20, "The timestamp with its KPI score & warning level", "-" * 20)
    display(generated_warnings)
```
