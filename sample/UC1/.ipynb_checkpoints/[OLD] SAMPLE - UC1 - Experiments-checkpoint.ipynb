```python
for inv_name in inv_names:
    df = inv_data[inv_name]
    
    # Compute the power --> P = V * A --> Power = Voltage * Ampere
    power = df["Vcc 1 (V)"] * df["Cc 1 (A)"]
    
    # Transform the WATT (W) into KILOWATT (kW)
    kilowatt = power / 1000
    df.insert(11, "DC Power (kW)", kilowatt)
    print(f"\n{inv_name}: Added the computed column 'DC Power (kW)' in the data (i.e., P = V *A ).")
```

```python
cols_to_drop = [
        "Allarme",
        "Humidity (%)",
        "Atmospheric Pressure (hPa)",
        "Rainfall (mm)",
        "Wind speed (m/s)",
        "Wind direction (째)"
    ] 
```

```python
col_dt = "Date/Time"
for inv_name in inv_names:
    print("\n"+30*"-", inv_name, 30*"-")
    df = inv_data[inv_name]
    
    # Drop the extra columns
    df.drop(columns = cols_to_drop, inplace=True)
    print(f"a) Some columns have been discarted (e.g., {cols_to_drop[0]}, {cols_to_drop[1]}, ...)")

    # Use the timestamp as index
    df.index = df[col_dt]
    df.drop(columns = col_dt, inplace=True)
    print("b) The timetamps are now used as index")

    # Drop timestamps duplicated (possible artefacts from the data cleaning)
    duplicated_ts = df[df.index.duplicated()].index
    df.drop(index = duplicated_ts, inplace= True)
    print(f"c) Dropped some artefacts (i.e., {len(duplicated_ts)} duplicated obs.)")
```

```python
var = "Irradiance (W/mq)"
var = "E. totale (kWh)"

# Create the figure (i.e., visual panel)
n_cols = 2 if len(inv_names) > 2 else 1 
n_rows = len(inv_names) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(10 * n_cols, 6 * n_rows), 
                         tight_layout= {"pad": 2}, facecolor = "white")
fig.suptitle(f"[{system_name}] " + r"$\bf{" + var.split("(")[0].replace(" ", "\\ ") + r"}$ (" + var.split("(")[1], fontsize=70, color='dimgray') #, y = 1.01

if axes.ndim == 1:
    axes = np.array([axes])

idk_row = 0
idk_column = 0    
for inv_name in inv_names:
    print("-" * 30 + f" {inv_name}" + "-" * 30)
    df = inv_data[inv_name]
    
    # DATA 
    timestamps = df.index.tolist()
    var_values = df[var].tolist()
    
    # MOVING AVERAGE (1 day, 1 week)
    var_values_moving_average_1d = df[var].rolling(window = 18, min_periods = 1).mean()    
    var_values_moving_average_1w = df[var].rolling(window = 18 * 7, min_periods=1).mean()
    threshold = 10**6
    problematic_range= var_values_moving_average_1d[var_values_moving_average_1d < threshold].index
    
    if len(problematic_range) > 0:
        print(f"PROBLEMATIC RANGE: FROM '{problematic_range[0]}' TO '{problematic_range[-1]}'")
    else:
        print("There is no problematic observations! Nice :)")
    
    graph = axes[idk_row, idk_column]
    sns.lineplot(x = timestamps, y = var_values, ax = graph, color= "orange",
                label = "Counter") 
    sns.lineplot(x = timestamps, y = var_values_moving_average_1d, ax = graph, color= "grey", 
                 label = "Moving Average [1 day]") 
    sns.lineplot(x = timestamps, y = var_values_moving_average_1w, ax = graph, color= "black",
                 label = "Moving Average [1 Week]")
    
    #axes.set_title("SOURCE: 'Solarimentro'")
    unit_measure = var.split("(")[1].rstrip(")")
    graph.set_title(f"Inverter N째{inv_name[-1]}", fontsize = 30, color= "dimgrey")
    graph.set_ylabel(unit_measure, fontsize = 40, color= "dimgrey")
    graph.set_xlabel("Date", fontsize = 25, color= "dimgrey")
    
    # Increment the row/column
    if idk_column + 1 < axes.shape[1]:
        idk_column += 1
    else:
        idk_column = 0
        idk_row += 1
        
plt.show()
```

```python
def analyse_missing_ts(df):
    # FOR DEBUG: Count the daily observations 
    df["date"] = pd.to_datetime(df.index).strftime('%Y-%m-%d')
    
    # Group observations by date
    grouped_df = df.groupby(by = ["date"]).count()[["DC Power (kW)", "Iac T (A)"]]
    grouped_df.rename(columns = {"DC Power (kW)": "Daily Observations"}, inplace=True)
    
    # Count the frequencies of the daily observations
    grouped_daily_obs_df = grouped_df.groupby("Daily Observations").count()
    grouped_daily_obs_df.rename(columns = {"Iac T (A)": "Frequency"}, inplace=True)
    grouped_daily_obs_df.sort_values(by = ["Frequency"], ascending=False, inplace=True)
    
    daily_obs = np.array(grouped_df["Daily Observations"].tolist())
    print("DAILY HOURLY OBSERVATIONS (Median): ", np.median(daily_obs))
    
    display(grouped_daily_obs_df["Frequency"].to_frame())

    # Grafico
    ax = sns.histplot(x = grouped_df.index, y = grouped_df["Daily Observations"], stat = "frequency") #, bins = 10
    ax.xaxis.set_major_locator(plt.MaxNLocator(12))
    ax.yaxis.set_major_locator(plt.MaxNLocator(20))
    plt.xticks(fontsize=12, rotation = 15)
    plt.yticks(fontsize=15)
    plt.rcParams.update({'figure.figsize': (15, 10)})
    plt.show()

    # Compute the missing hours
    dates = grouped_df.index.tolist()
    for date in dates:
        daily_ts = pd.to_datetime(df.loc[date, :].index)
        daily_hours = [int(ts.strftime("%H")) for ts in daily_ts]
        grouped_df.loc[date, "First Hour"] = int(daily_hours[0])
        grouped_df.loc[date, "Last Hour"] = int(daily_hours[-1])

        missing_hours_idk = [item[0] for item in np.argwhere(np.diff(daily_hours) > 1)]
        periods = []
        for hour_idk in missing_hours_idk:
            period = (daily_hours[hour_idk], daily_hours[hour_idk + 1], daily_hours[hour_idk + 1] - daily_hours[hour_idk])
            periods.append(period)
        grouped_df.loc[date, "Missed central hours"] = ", ".join(str(period[0]) + ":" 
                                                                 + str(period[1]) + f" (gap: {period[2]} hours)" 
                                                                 for period in periods)
    
    # Group furthermore
    advanced_grouping = grouped_df.groupby(by = ["First Hour", "Last Hour"]).count()["Daily Observations"].to_frame()
    advanced_grouping.sort_values(by = ["Daily Observations"], ascending=False, inplace=True)
    display(advanced_grouping)
    
    cond = grouped_df["Missed central hours"] != "" 
    cond2 = grouped_df["Missed central hours"].str.contains("0:5|0:6") == False
    problematic_missing_hours = grouped_df[cond & cond2]
    print("-" * 40 + " PROBLEMATIC GAPS " + "-" * 40)
    print("Missing night periods have been removed ('0:5' & '0:6')")
    print(f"Problematic days with missing hours: {len(problematic_missing_hours)} day(s) out of {len(grouped_df)} days " \
          f"({round(len(problematic_missing_hours)/len(grouped_df)*100, 2)} %)")
    display(grouped_df.drop(columns = "Iac T (A)")[cond & cond2])
    
    df.drop(columns = "date", inplace = True)
    
```
